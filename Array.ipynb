```python
from array import *
```

```python
arr1=array('i',[2,4,6,8,10])
print(arr1)
```

```python
arr2=array('d',[2,4,6,8,10])
print(arr2)
```

### Insert Elements

Time Complexity of Insertion at the end =  O(1)

Time Complexity of Insertion at the beginning=  O(n)      because move 1 index & insert



```python
arr1.append(12)   #end
arr1
```

```python
arr1.insert(2,68)  #given index
arr1
```

```python
arr1.extend([30,40,50])
arr1
```

### Travelsal

Time Complexity = O(n)



```python
for i in arr1:
    print(i)
```

### Access array

Time Complexity = O(1)

```python
arr1[4]
```

### Delete Elements

Time Complexity of Insertion at the end = O(1)

Time Complexity of Insertion at the beginning= O(n) because 1 space left after remove



```python
arr1.remove(4)  #given element
arr1
```

```python
arr1.pop()  #last element
arr1
```
